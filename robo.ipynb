```python
import yfinance as yf
import pandas as pd
import time

# config do robo
TICKER = "AAPL"  # ação acompanhar (apple)
SHORT_WINDOW = 20  #janela da média movel curta
LONG_WINDOW = 50   # janela da média movel longa
CAPITAL_INICIAL = 10000  # capital inicial em dólares
POSICAO = 0  # numero de ações compradas

def obter_dados(ticker):
    """dados historicos do Yahoo finance."""
    dados = yf.download(ticker, period="6mo", interval="1d")
    return dados

def calcular_medias_moveis(dados):
    """dalcula médias móveis curtas e longas."""
    dados['SMA20'] = dados['Close'].rolling(window=SHORT_WINDOW).mean()
    dados['SMA50'] = dados['Close'].rolling(window=LONG_WINDOW).mean()
    return dados

def tomar_decisao(dados, capital, posicao):
    """decide comprar, vender ou manter com base em médias móveis."""
    ultimo_preco = dados['Close'].iloc[-1]
    sma20 = dados['SMA20'].iloc[-1]
    sma50 = dados['SMA50'].iloc[-1]

    if sma20 > sma50 and posicao == 0:  # Compra
        acoes_compradas = capital // ultimo_preco
        capital_restante = capital - (acoes_compradas * ultimo_preco)
        # Converte ultimo_preco e capital_restante to float before formatting
        print(f"Comprado {acoes_compradas} ações a {float(ultimo_preco):.2f}. Capital restante: {float(capital_restante):.2f}")
        return capital_restante, acoes_compradas
    elif sma20 < sma50 and posicao > 0:  # Venda
        capital_venda = posicao * ultimo_preco
        # converte ultimo_preco e capital_venda to float before formatting
        print(f"Vendido {posicao} ações a {float(ultimo_preco):.2f}. Capital total: {float(capital_venda):.2f}")
        return capital_venda, 0
    else:  # manter
        # aguarda a melhor oportunidade
        print(f"Nenhuma ação tomada. Preço atual: {float(ultimo_preco):.2f}")
        return capital, posicao

# execução do robo
if __name__ == "__main__":
    capital = CAPITAL_INICIAL
    posicao = POSICAO
    while True:
        print("\nObtendo dados...")
        dados = obter_dados(TICKER)
        dados = calcular_medias_moveis(dados)

        capital, posicao = tomar_decisao(dados, capital, posicao)


        time.sleep(3600)  # atualiza a cada hora
```
